```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
initial_df = pd.read_csv("data/Crime_Reports.csv")
initial_df
```

```python
initial_df.columns
```

```python
columns_to_keep = ['Highest Offense Description', 'Family Violence', 'Occurred Date', 'Latitude', 'Longitude']
second_df = initial_df[columns_to_keep].copy()

second_df
```

```python
third_df = second_df[second_df["Family Violence"] != "Y"]
third_df.reset_index(drop=True , inplace=True)
third_df
```

```python
columns_to_keep2 = ['Highest Offense Description', 'Occurred Date', 'Latitude', 'Longitude']
fourth_df = third_df[columns_to_keep2].copy()
fourth_df
```

```python
fourth_df["year"] = fourth_df["Occurred Date"].str[-4:]
fourth_df.drop(columns=["Occurred Date"], inplace=True)
fourth_df
```

```python

year2020 = "2020"
year2021 = "2021"
year2022 = "2022"
year2023 = "2023"
year2024 = "2024"
```

```python
df_2020 = fourth_df[fourth_df["year"] == year2020].reset_index(drop=True)
df_2020
```

```python
df_2021 = fourth_df[fourth_df["year"] == year2021].reset_index(drop=True)
df_2021
```

```python
df_2022 = fourth_df[fourth_df["year"] == year2022].reset_index(drop=True)
df_2022
```

```python
df_2023 =  fourth_df[fourth_df["year"] == year2023].reset_index(drop=True)
df_2023
```

```python
df_2024 = fourth_df[fourth_df["year"] == year2024].reset_index(drop=True)
df_2024
```

```python
df_2020.to_csv("Austin_crime_reports_2020.csv", index=False)
df_2021.to_csv("Austin_crime_reports_2021.csv", index=False)
df_2022.to_csv("Austin_crime_reports_2022.csv", index=False)
df_2023.to_csv("Austin_crime_reports_2023.csv", index=False)
df_2024.to_csv("Austin_crime_reports_2024.csv", index=False)
```
